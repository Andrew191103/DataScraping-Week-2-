```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_best-selling_video_games'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
#print(soup)
```

```python
table = soup.find_all('table')[1]
```

```python
#soup.find('table', class_= 'wikitable plainrowheaders sortable')
```

```python
print(table)
```

```python
soup.find_all('th')
```

```python
world_titles = table.find_all('th')
```

```python
world_titles
```

```python
world_table_titles = [title.text for title in world_titles]

print (world_table_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = world_table_titles)

df
```

```python
for row in table.find_all('tr')[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    
    # Fill in missing values with empty strings to match column count
    while len(individual_row_data) < len(world_table_titles):
        individual_row_data.append('')
    
    # Add the row to the DataFrame
    df.loc[len(df)] = individual_row_data
    
print(df)
```
